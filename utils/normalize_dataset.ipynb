```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

```

```python
FILE = '../datasets/custom-diabetes.csv'

def load_data(file=FILE):
    return pd.read_csv(file)

data = load_data()
```

```python
labels = data["class"].copy()
data = data.drop("class", axis=1)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
```

```python
scaled_df = scaler.fit_transform(data)
scaled_df
```

```python
scaled_df = pd.DataFrame(scaled_df, columns=['0', '0', '0','0', '0', '0','0', '0'])
```

```python
type(labels)
labels = pd.DataFrame(labels)
# scaled_df.join(labels)
result = pd.concat([scaled_df, labels], axis=1, join='inner')
print(labels.shape)
result.head(10)
```
